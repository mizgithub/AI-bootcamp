## 1. Exploring data
`About the dataset`: This dataset is originally from the National Institute of Diabetes and Digestive and Kidney Diseases. The objective of the dataset is to diagnostically predict whether or not a patient has diabetes, based on certain diagnostic measurements included in the dataset. Several constraints were placed on the selection of these instances from a larger database. In particular, all patients here are females at least 21 years old of Pima Indian heritage.

The datasets consists of several medical predictor variables and one target variable, Outcome. Predictor variables includes the number of pregnancies the patient has had, their BMI, insulin level, age, and so on.

## Handling Missing Value


### Checking existence of missing value 
        `nan` or `na` values

```python
dataset.isna() ## Display True if it is a missing value and False if it is not
```

```python
dataset.isna().sum() ## display the total number of missing values per column
```

```python
## ploting a graph to visualiize missing values ratio 
dataset.isna().sum().plot(kind='bar')
```

### Handling missing value by droping (removing)

```python
## Drop rows that contain a missing value
modified_dataset = dataset.dropna() 

## see the difference 
print(len(modified_dataset))
print(len(dataset))
```

```python
## Drop a column that contain a missing value
modified_dataset = dataset.dropna(axis =1)
## see the difference 
print(len(modified_dataset))
print(len(dataset))
print(modified_dataset.columns)
print(dataset.columns)
```

```python
## Drop a column or row that contain a missing values above a specified threshold
modified_dataset1 = dataset.dropna(thresh=8) ## keep rows that has a t least 8 non-missing value
modified_dataset2 = dataset.dropna(thresh = 760, axis=1) # keep columns that has a leats 760 non-missing values
## see the difference 
print(len(modified_dataset1))
print(len(dataset))
print(modified_dataset2.columns)
print(dataset.columns)
```

### Fill missing value with a specified constant value

```python
modified_dataset = dataset.fillna(value=0)
## see the difference
modified_dataset.notna().sum().plot(kind='bar') ## No missing value 
```

### Filling missing value with estimated or predicted value (Imputation)
`SimpleImputer` from `scikit-learn` is a popular choice for imputing missing values

package installation: `pip install scikit-learn`

Importing: ` from sklearn.impute import SimpleImputer`

The SimpleImputer class provides basic strategies for imputing missing values. Missing values can be imputed with a provided constant value, or using the statistics (mean, median or most frequent) of each column in which the missing values are located. This class also allows for different missing values encodings.

```python
from sklearn.impute import SimpleImputer
import numpy as np
```

```python
data = dataset.copy()
imp = SimpleImputer(missing_values=np.nan,strategy='mean')
imp.fit(data)
data = imp.transform(data) ## imputer returns ndarray therfore you need to convert it into dataframe
data = pd.DataFrame(data) ## bulding dataframe from ndarray
```

```python
data.notna().sum().plot(kind='bar')
```

Coorelation analysis
A correlation of -1 shows a negative correlation,
A correlation of +1 shows a positive correlation.
A correlation of 0 indicates that there is no relationship between the variables.
Example: using pearson correlation 
apply Pearson's correlation between all variables in the dataset `data.corr(method = 'pearson')`

```python
dataset.corr(method='pearson')
```

```python
dataset.corr(method='pearson').plot()
```

```python
# Correlation matrix with variable names
correlations = dataset.corr()
columns = dataset.columns
import numpy as np       
fig = plt.figure()       
ax = fig.add_subplot(111)    
cax = ax.matshow (correlations, vmin = -1, vmax = 1) 
fig.colorbar(cax)     

#The array defines the size of the 9x9 square to be plotted
ticks = np.arange(0, 9, 1) 

# Take the size of "ticks" and place it on the x axis
ax.set_xticks(ticks) 

# Take the size of "ticks" and place it on the
ax.set_yticks(ticks)

# Apply the columns listed at the beginning as labels
ax.set_xticklabels(columns) 

# Applies the columns listed at the beginning as labels
ax.set_yticklabels(columns) 
plt.show() # Plot
```

## Visualizing the data



## Histogram
With the histogram, we can quickly assess the distribution of each attribute. Histograms group data into bins and provide a count of the number of observations in each bin.

We can quickly check the dataâ€™s symmetry with the histogram and whether it is in a normal distribution. It will help to identify outliers.

```python
dataset.hist()
plt.show()
```

### Density plot
Density Plots are another way to visualize the distribution of data for each attribute. The plot is like a kind of abstract histogram with a smooth curve through the top of a histogramâ€™s bins.
